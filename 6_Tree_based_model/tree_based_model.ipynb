```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pylab import style
style.use('ggplot')   #图画风格
plt.rcParams['font.sans-serif'] = ['SimHei']  #字体风格
plt.rcParams['axes.unicode_minus'] = False    #是否中文字体变小
plt.rcParams['figure.figsize'] = (12,7)       #图像大小

import os,sys
sys.path.append(os.path.abspath("../"))
from util import util
```

```python
from xgboost import XGBRegressor
from lightgbm.sklearn import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import GridSearchCV,train_test_split,KFold
from sklearn.tree import ExtraTreeRegressor,DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor,BaggingRegressor,VotingRegressor,ExtraTreesRegressor,RandomForestRegressor
```

```python
def plot_result(yhat,y,test_year='2016'):
    date = pd.date_range(start=f'{test_year}/01/01 0:00:00',periods=len(yhat),freq='h')
    plt.figure(figsize=(30,7))
    plt.plot(date,yhat)
    plt.plot(date,y)
    mse = util.mse(yhat,y)
    mape = util.mape(yhat,y)
    plt.title(f'mse: {mse}, mape: {mape}')
    plt.legend(('predict','real'))
    plt.show()
```

```python
train_file_name = '../input/years/2015_year.csv'  
test_file_name = '../input/years/2016_year.csv'  
train_df = util.load_data(file_name=train_file_name,start_point='2015/1/1 0:00:00')
test_df = util.load_data(file_name=test_file_name,start_point='2016/1/1 0:00:00')
train_df = util.whole_flow(train_df)
test_df = util.whole_flow(test_df)

train_df = util.create_shift_features(train_df,shift_columns=util.columns,offsets=[0,0,59,59],periods=[24]*4)
test_df = util.create_shift_features(test_df,shift_columns=util.columns,offsets=[0,0,59,59],periods=[24]*4)
```

```python
features = train_df.corrwith(train_df['zhexi_in']).abs().sort_values(ascending=False)[:40].index.values
scaler = MinMaxScaler()
train_arr = scaler.fit_transform(train_df[features])
test_arr = scaler.transform(test_df[features])
# train_arr = np.array(train_df[features])
# test_arr = np.array(test_df[features])
train_arr.shape,test_arr.shape
```

```python
X_train = train_arr[:,1:]
y_train = train_arr[:,0]
X_test = test_arr[:,1:]
y_test = test_arr[:,0]
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

# Catboost

```python
param_grid = {'n_estimators':[i for i in range(50,500,50)]}
other_params = {'learning_rate': 0.1,'n_estimators':1000,'l2_leaf_reg':3,'depth':6,
                #'border_count':32,
                'task_type':'GPU','min_child_samples':0.3,'verbose':0}
estm_grad = GridSearchCV(CatBoostRegressor(**other_params),
                        param_grid=param_grid,
                        scoring='neg_mean_squared_error')
print(estm_grad.fit(X_train,y_train))
print(estm_grad.best_params_,estm_grad.best_score_)
other_params['n_estimators'] = estm_grad.best_params_['n_estimators']
```

```python
param_grid = {'l2_leaf_reg':[2,3,4,7,11]}
estm_grad = GridSearchCV(CatBoostRegressor(**other_params),
                        param_grid=param_grid,
                        scoring='neg_mean_squared_error')
print(estm_grad.fit(X_train,y_train))
print(estm_grad.best_params_,estm_grad.best_score_)
other_params['l2_leaf_reg'] = estm_grad.best_params_['l2_leaf_reg']
```

```python
param_grid = {'depth':[2,3,4,5,6],'min_child_samples':np.linspace(.3,1.0,8)}
estm_grad = GridSearchCV(CatBoostRegressor(**other_params),
                        param_grid=param_grid,
                        scoring='neg_mean_squared_error')
print(estm_grad.fit(X_train,y_train))
print(estm_grad.best_params_,estm_grad.best_score_)
other_params['depth'] = estm_grad.best_params_['depth']
other_params['min_child_samples'] = estm_grad.best_params_['min_child_samples']
```

```python
param_grid = {'learning_rate':[0.001,0.01,0.1,0.2]}
estm_grad = GridSearchCV(CatBoostRegressor(**other_params),
                        param_grid=param_grid,
                        scoring='neg_mean_squared_error')
print(estm_grad.fit(X_train,y_train))
print(estm_grad.best_params_,estm_grad.best_score_)
other_params['learning_rate'] = estm_grad.best_params_['learning_rate']
```

```python
estm_cb = CatBoostRegressor(**other_params)
util.plot_learning_curve(estm_cb,f"CB({str(other_params)})", X_train, y_train,train_sizes=np.linspace(.25, 1.0, 20),cv=10,scoring='neg_mean_squared_error',n_jobs=14)
```

```python
estm_cb = CatBoostRegressor(**other_params)
estm_cb.fit(X_train,y_train)
yhat = estm_cb.predict(X_test)
```

```python
plot_result(yhat,y_test)
```

# Adaboost

```python
param_grid = {'criterion': ['mse'], 
              'min_impurity_decrease': np.linspace(0, 1, 100),
              'max_depth': np.arange(2,10),
              'min_samples_split': np.arange(2,30,2)}
estm_dtree = GridSearchCV(DecisionTreeRegressor(),param_grid,scoring='neg_mean_squared_error',n_jobs = 15)
print(estm_dtree.fit(X_train,y_train))
print(estm_dtree.best_params_,estm_dtree.best_score_)
```

```python
estm_dtree = DecisionTreeRegressor(criterion= 'mse', max_depth=7, min_impurity_decrease= 0.0, min_samples_split= 6)
param_grid = {'n_estimators':[28],
              'loss':['square'],
              'learning_rate':np.linspace(0.05,0.3,100),
             }
estm_ada = GridSearchCV(AdaBoostRegressor(base_estimator=estm_dtree),
                        param_grid,scoring='neg_mean_squared_error',n_jobs = 15)
print(estm_ada.fit(X_train,y_train))
print(estm_ada.best_params_,estm_ada.best_score_)
```
